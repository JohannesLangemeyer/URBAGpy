```python
#importem la llibreria pandas i la matplotlib
import pandas as pd
import matplotlib.pyplot as plt
# https://matplotlib.org/gallery/index.html
```

```python
#importem les dades ara ja del fitxer csv
metData = pd.read_csv('../Data/allData.csv')
metData
```

```python
#filtrem les dades per obtenir les de gener de 2018
Jan2018 = metData[metData['DATA'] < '2018-02-01 00:00:00']
Jan2018
```

```python
#ordenem les dades per data
Jan2018.sort_values(by=['DATA'])
```

```python
# comprovem quants valors diverents conté el camp ESTACIO
Jan2018['ESTACIO'].unique()
```

```python
#extraiem les dades de temperatura de D5
D5T = Jan2018[Jan2018['ESTACIO']=='D5']['T']
```

```python
#extraiem les dades de temperatura de WU
WUT = Jan2018[Jan2018['ESTACIO']=='D5']['T']
```

```python
# Extraiem els timepoints per a l'eix de les x's
time = Jan2018['DATA'].unique()
```

```python
# dibuixem el gràfic, veiem que té moltes dades i potser no es tot lo informatiu que voldriem
plt.plot(time, D5T, label='D5-Fabra')
plt.plot(time, WUT, label='WU-Badalona')
plt.title("Temperature")
plt.legend()
plt.show()
```

```python
# Ens plantegem que volem graficar la temperatura mínima diaria del més de gener de cada observatori
```

```python
time = []
TminD5 = []
TminWU = []
TminX2 = []
TminX4 = []
TminX8 = []

Jan2018 = metData[metData['DATA'] < '2018-02-01 00:00:00']
for i in range(1,32):
    time.append(i)
    dayData = Jan2018[ (Jan2018['DATA'] > '2018-01-'+str(i).zfill(2)+' 00:00:00') & (Jan2018['DATA'] < '2018-01-'+str(i).zfill(2)+' 23:59:59')]
    TminD5.append( dayData[dayData['ESTACIO'] == 'D5']['T'].min() )
    TminWU.append( dayData[dayData['ESTACIO'] == 'WU']['T'].min() )
    TminX2.append( dayData[dayData['ESTACIO'] == 'X2']['T'].min() )
    TminX4.append( dayData[dayData['ESTACIO'] == 'X4']['T'].min() )
    TminX8.append( dayData[dayData['ESTACIO'] == 'X8']['T'].min() )
    
```

```python
# dibuixem el gràfic, veiem que té moltes dades i potser no es tot lo informatiu que voldriem
plt.plot(time, TminD5, label='D5-Fabra')
plt.plot(time, TminWU, label='WU-Badalona')
plt.plot(time, TminX2, label='WU-Zoo')
plt.plot(time, TminX4, label='WU-Raval')
plt.plot(time, TminX8, label='WU-ZonaUni')

plt.title("Temperature")
plt.legend()
plt.show()
```
