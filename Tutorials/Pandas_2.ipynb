<h2> Data manipulation</h2>

```python
#importem la llibreria pandas
import pandas as pd
```

```python
#carreguem tots els fitxers
D5 = pd.read_excel('../Data/Dades_4298_D5_10.xlsx')
WU = pd.read_excel('../Data/Dades_4298_WU_6.xlsx')
X2 = pd.read_excel('../Data/Dades_4298_X2_.xlsx')
X4 = pd.read_excel('../Data/Dades_4298_X4_10.xlsx')
X8 = pd.read_excel('../Data/Dades_4298_X8_10.xlsx')
Y7 = pd.read_excel('../Data/Dades_4298B_Y7.xlsx')
```

```python
#comprovem les dimensions de cada data frame
D5.shape, WU.shape, X2.shape, X4.shape, X8.shape, Y7.shape
```

```python
#com que y7 es diferent ens el mirem, i veiem que no té alguns camps
Y7
```

```python
# llistem tots els caps de cada data frme per veure si hi ha alguna altre discrepància
ColSumArray = [list(D5),list(WU),list(X2),list(X4),list(X8),list(Y7)]
ColSum = pd.DataFrame(ColSumArray)
ColSum
```

```python
# Creem un data frame amb les dades de totes les taules
# ignore_index=True serveix per dir-li que no faci cas del index originals i que els reindexi
AllData = pd.concat([D5,WU,X2,X4,X8], ignore_index=True)
AllData
```

```python
# Ara guardem el dataframe en un fitxer allData.csv
# index = False serveix per dir-li que no volem un camp amb l'index
AllData.to_csv('../Data/allData.csv', index = False)
```
